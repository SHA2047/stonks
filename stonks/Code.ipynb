```python
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta
import os
from pathlib import Path
from tqdm import tqdm
```

```python
path = str(Path(os.getcwd()).parent)
# List of stock tickers - You need to fill this with the tickers you want
tickers = pd.read_csv(path + "/configs/EQUITY_L.csv")["SYMBOL"].tolist()#['SJVN.NS']#, 'TCS.NS', 'INFY.BO', 'HDFCBANK.BO'] # Example tickers

end_date = datetime.today()#.strftime("%Y-%m-%d")
start_date = datetime.today() - timedelta(days = 2)

# Downloading data
def download_data(tickers, start_date, end_date):
    stock_data = {}
    for ticker in tqdm(tickers):
        stock = yf.Ticker(ticker + ".ns")
        data = stock.history(start=start_date, end=end_date)
        stock_data[ticker] = data
    return stock_data

# Saving data
stock_data = download_data(tickers, start_date, end_date)
alldata = pd.DataFrame()
for ticker, data in stock_data.items():
    temp = data
    temp['ticker'] = ticker
    alldata = pd.concat([temp, alldata], axis = 0)
alldata.reset_index(inplace = True)

print("Data download complete.")
```
